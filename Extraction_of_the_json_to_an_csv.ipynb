```python
import pandas as pd
import json


with open('/content/drive/My Drive/Procesamiento de Datos/Parcial/season_stats.json') as file:
    json_data = json.load(file)
teams_list = []
aggregate_stats_list = []
players_list = []

for match_id, teams in json_data.items():
    for team_id, team_info in teams.items():
        team_details = team_info['team_details']
        team_details['match_id'] = match_id
        team_details['team_id'] = team_id
        teams_list.append(team_details)
        aggregate_stats = team_info['aggregate_stats']
        aggregate_stats['match_id'] = match_id
        aggregate_stats['team_id'] = team_id
        aggregate_stats_list.append(aggregate_stats)
        for player_name, player_info in team_info['Player_stats'].items():
            player_stats = player_info['Match_stats']
            player_stats.update(player_info['player_details'])
            player_stats['match_id'] = match_id
            player_stats['team_id'] = team_id
            player_stats['player_name'] = player_name
            players_list.append(player_stats)

df_team_details = pd.DataFrame(teams_list)
df_aggregate_stats = pd.DataFrame(aggregate_stats_list)
df_player_stats = pd.DataFrame(players_list)
```
